#Preprocessing

```python
import ast
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#taking out the keywords
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

###Cleaning Genres

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#Genres cleaned
```

###Cleaning Keywords

```python
movies['keywords'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#Keywords cleaned
```

###Cleaning Cast

```python
#taking only first three actors
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#Casts cleaned
```

###Cleaning Crew

```python
#taking out the director from crew
def fetchdirector(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'].apply(fetchdirector)
```

```python
movies['crew'] = movies['crew'].apply(fetchdirector)
```

```python
movies.head()
```

```python
#crew cleaned
```

###Splitting overview

```python
movies['overview'][0]
```

```python
#changing overview from string to tags
movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

###Removing Spaces

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies
```

###Creating Tags column

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

###Converting tags to lowercase

```python
newmovies.head()
```

###Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
newmovies['tags'].apply(stem)
```

```python
newmovies['tags'] = newmovies['tags'].apply(stem)
```

#Vectorization : Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')#removing stop words like and of from to etc.
```

```python
vectors = cv.fit_transform(newmovies['tags']).toarray()
vectors
```
